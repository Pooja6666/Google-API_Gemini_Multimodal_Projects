## Audio to text by using gemini

```python
#audio_file = genai.upload_file(path=r"C:\Users\DELL\Downloads\Calm Down Calm Down(PagalWorld.com.so).mp3")
import pathlib

# Initialize a Gemini model appropriate for your use case.
model = genai.GenerativeModel('models/gemini-1.5-flash')

# Create the prompt.
prompt = "generate full lyrics."

# Load the samplesmall.mp3 file into a Python Blob object containing the audio
# file's bytes and then pass the prompt and the audio to Gemini.
response = model.generate_content([
    prompt,
    {
        "mime_type": "audio/mp3",
        "data": pathlib.Path(r"C:\Users\DELL\Downloads\Calm Down Calm Down(PagalWorld.com.so).mp3").read_bytes()
    }
])

# Output Gemini's response to the prompt and the inline audio.
print(response.text)
```
